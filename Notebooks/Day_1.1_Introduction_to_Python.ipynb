# Variables

Variables are words or symbols that are used to represent some value. They're kind of like boxes that you can store things in. Variables in Python are used very similarly to variables that you have probably seen in math class. For example, if you assign x = 5 and compute 9 + x = _ ?

Exactly - 14! Let's see some examples of how this works in code.



```python
# assign the value of 6*13 to the variable result1
result1 = 6*13 ##

# assign the value of 98/4 to the variable result2
result2 = 98/4 ##
```

You might notice that after running the above cell, nothing was printed to the screen. This is because our code commanded Python only to assign the values to the variable names, not to print to the screen. This is a good reminder that Python is very literal, and will only do **exactly what we command it to do**.

If we want to see the values assigned to each variable, we can use the `print` command we learned above to print them out.

```python
# print the value stored in the variable result1
print(result1) ##

# print the value stored in the variable result2
print(result2) ##
```

Cool! Now we've seen that the `print` function in Python can accept different inputs, such as a string (as in `"Hello World"`), or a variable name (as in `result1`). We see that when we use the variable name `result1`, it prints the *value* of that variable, not the variable name itself. We will talk more about how we differentiate strings from variable names in the next lesson.

## Updating Variables

Variables are very helpful in coding. Not only do they prevent us from having to remember and type the values over and over again, it allows us to update certain values when we have new information. For example, let's say we have a variable `apples` that stores the number of apples we have in our refrigerator.

```python
apples = 2
```

Now let's say I went to the store and bought half a dozen apples. Now I want to update the value of `apples` without having to remember how many we started with. How should I do this?

```python
# add 6 to apples
apples + 6 ##

# print apples
print(apples) ##
```

Did you get the value you expected?

```python
# add 6 to apples and update apples
apples += 6

# print apples
print(apples)
```

Notice that we can't just add 6 to `apples` and have Python know we mean for it to save that result, **we need to explicitly tell Python to update the value of** `apples`. To do this we use the `+=` operator. This adds 6 to the previous value of `apples` (2) and updates the variable `apples` to store this new value. This is called **incrementing**.

**Exercise:** Let's say I had some friends coming over for dinner and I wanted to make an apple pie for dessert. This used up 4 apples. How do you think we should update the variable `apples` to reflect the number of apples that remain in the reifrigerator?

```python
# subtract 4 from apples and update it
apples -= 4

# print apples
print(apples)
```

This is an example of **decrementing**.

Can you think of another way to increment or decrement without using the above operators?

```python
# add 2 to apples and update it
apples = apples + 2

# print
print(apples)
```

## Variable Naming
So far, we've been creating variable names that are only one letter or word, but as you create more variables, you may want to create variable names that are multiple words and are more descriptive.

There are two variable name formats that are widely recommended in the python community: "snake case" and "camel case". Snake case is when you delimit words by underscores. Camel case is when you capitalize the first letter in every word after the first word.

While other naming schemes won't necessarily throw errors, you'll find that as you continue learning to code, that these naming schemes are generally reserved for certain types of variables, so it's best to avoid using them to avoid confusion. 

Two absolute requirements of variable naming are that: (1) there can not be a space in your variable name and (2) your variable name can not start with a number.

```python
# Good

my_name = "Grace"     # snake case
myName = "Grace"      # camel case

# Bad
#MY_NAME = "Grace"
#MyName = "Grace"
#1result = "Grace"
```

One last thing to note when chosing variable names is you want to try not to use variable names for commands or functions that already exist in the python language. For example, let's see what happens when we use **print** as a variable name.

```python
# assign the number 1 to the variable name print
print = 1 ##
```

```python
# now try printing a hello message again
print("Hello") ##
```

You should get an error message here. This is because we have overwritten the print command with our variable. When possible, we want to avoid doing this, but if it happens, here's how you fix it to get the print statement back.

```python
# delete our print variable
del print ##

# print our greeting again
print("Hello") ##
```
